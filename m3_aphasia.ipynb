```python
import pandas as pd
import numpy as np
import os
import glob
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
from sklearn.ensemble import RandomForestClassifier
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, LSTM, GRU, Conv1D, Conv2D, MaxPooling1D, MaxPooling2D
from tensorflow.keras.layers import Dropout, BatchNormalization, GlobalAveragePooling1D, GlobalAveragePooling2D
from tensorflow.keras.layers import Input, Reshape, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc
import warnings
warnings.filterwarnings('ignore')

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

class AphasiaDetectionSystem:
    """
    End-to-end Aphasia Detection System based on the research paper
    """

    def __init__(self, data_path):
        """
        Initialize the Aphasia Detection System

        Args:
            data_path (str): Path to the folder containing CSV files
        """
        self.data_path = data_path
        self.data = None
        self.X = None
        self.y = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()

        # Feature columns based on the CSV structure
        self.acoustic_features = [
            'mfcc_mean', 'mfcc_std', 'zcr_mean', 'rms_mean', 'pitch_mean',
            'jitter', 'shimmer', 'formant1', 'formant2', 'formant3'
        ]

        self.linguistic_features = [
            'total_words', 'unique_words', 'ttr', 'avg_word_length',
            'speech_rate_wps', 'filler_count', 'function_word_count',
            'noun_count', 'verb_count', 'pronoun_count'
        ]

        self.temporal_features = [
            'duration_sec', 'pause_before_sec'
        ]

        # All features combined
        self.all_features = self.acoustic_features + self.linguistic_features + self.temporal_features

    def load_data(self):
        """
        Load all CSV files from the specified directory and combine them
        """
        print("Loading data from CSV files...")

        # Get all CSV files from the directory
        csv_files = glob.glob(os.path.join(self.data_path, "*.csv"))
        print(f"Found {len(csv_files)} CSV files")

        # Load and combine all CSV files
        dataframes = []
        for file in csv_files:
            try:
                df = pd.read_csv(file)
                dataframes.append(df)
                print(f"Loaded {file}: {df.shape}")
            except Exception as e:
                print(f"Error loading {file}: {e}")

        # Combine all dataframes
        if dataframes:
            self.data = pd.concat(dataframes, ignore_index=True)
            print(f"Combined dataset shape: {self.data.shape}")

            # Display basic information about the dataset
            print("\nDataset Info:")
            print(self.data['speaker_type'].value_counts())
            print(f"Total patients: {len(self.data[self.data['speaker_type'] == 'Patient'])}")
            print(f"Total controls: {len(self.data[self.data['speaker_type'] == 'Control'])}")

        else:
            raise Exception("No valid CSV files found!")

    def preprocess_data(self):
        """
        Preprocess the data for model training
        """
        print("Preprocessing data...")

        if self.data is None:
            self.load_data()

        # Filter only patient data (as per paper methodology)
        patient_data = self.data[self.data['speaker_type'] == 'Patient'].copy()

        # Create binary classification based on AQ score
        # Using AQ >= 90 as High-AQ (similar to paper's approach)
        # Adjust threshold based on your data distribution
        aq_threshold = 90 #patient_data['AQ_score'].median()  # Use median as threshold
        print(f"Using AQ threshold: {aq_threshold}")

        patient_data['aphasia_severity'] = (patient_data['AQ_score'] < aq_threshold).astype(int)
        print(f"Severe aphasia (1): {sum(patient_data['aphasia_severity'])}")
        print(f"Mild aphasia (0): {len(patient_data) - sum(patient_data['aphasia_severity'])}")

        # Handle missing values
        for feature in self.all_features:
            if feature in patient_data.columns:
                patient_data[feature] = patient_data[feature].fillna(patient_data[feature].median())

        # Prepare features and target
        self.X = patient_data[self.all_features].copy()
        self.y = patient_data['aphasia_severity'].copy()

        # Remove any rows with remaining NaN values
        mask = ~(self.X.isnull().any(axis=1) | self.y.isnull())
        self.X = self.X[mask]
        self.y = self.y[mask]

        print(f"Final dataset shape: {self.X.shape}")
        print(f"Feature distribution - Severe: {sum(self.y)}, Mild: {len(self.y) - sum(self.y)}")

        # Scale features
        self.X_scaled = self.scaler.fit_transform(self.X)

    def split_data(self, test_size=0.2, val_size=0.2, random_state=42):
        """
        Split data into train, validation, and test sets

        Args:
            test_size (float): Proportion of data for testing (0.2 = 20%)
            val_size (float): Proportion of training data for validation (0.2 = 20%)
            random_state (int): Random state for reproducibility
        """
        print("Splitting data...")

        # First split: separate test set (20%)
        X_temp, self.X_test, y_temp, self.y_test = train_test_split(
            self.X_scaled, self.y, test_size=test_size, random_state=random_state, stratify=self.y
        )

        # Second split: separate validation set from remaining data (20% of original 80%)
        self.X_train, self.X_val, self.y_train, self.y_val = train_test_split(
            X_temp, y_temp, test_size=val_size, random_state=random_state, stratify=y_temp
        )

        print(f"Training set: {self.X_train.shape[0]} samples")
        print(f"Validation set: {self.X_val.shape[0]} samples")
        print(f"Test set: {self.X_test.shape[0]} samples")

        print(f"Training distribution - Severe: {sum(self.y_train)}, Mild: {len(self.y_train) - sum(self.y_train)}")
        print(f"Validation distribution - Severe: {sum(self.y_val)}, Mild: {len(self.y_val) - sum(self.y_val)}")
        print(f"Test distribution - Severe: {sum(self.y_test)}, Mild: {len(self.y_test) - sum(self.y_test)}")

    def create_cnn_model(self, input_shape):
        """
        Create CNN model similar to the paper

        Args:
            input_shape (tuple): Shape of input features
        """
        model = Sequential([
            # Reshape input for CNN (treating features as 1D sequence)
            Input(shape=input_shape),
            Reshape((input_shape[0], 1)),

            # Convolutional layers
            Conv1D(64, 3, activation='relu', padding='same'),
            BatchNormalization(),
            MaxPooling1D(2),

            Conv1D(128, 3, activation='relu', padding='same'),
            BatchNormalization(),
            MaxPooling1D(2),

            Conv1D(256, 3, activation='relu', padding='same'),
            BatchNormalization(),

            # Global Average Pooling (as mentioned in paper)
            GlobalAveragePooling1D(),

            # Dense layers
            Dense(128, activation='relu'),
            Dropout(0.5),
            Dense(64, activation='relu'),
            Dropout(0.3),

            # Output layer (binary classification)
            Dense(1, activation='sigmoid')
        ])

        return model

    def create_gru_model(self, input_shape):
        """
        Create sequence-to-one GRU-RNN model as described in the paper

        Args:
            input_shape (tuple): Shape of input features
        """
        model = Sequential([
            Input(shape=input_shape),
            Reshape((input_shape[0], 1)),

            # GRU layers (as per paper: 2 uni-directional GRU layers with 200 hidden units each)
            GRU(200, return_sequences=True),
            Dropout(0.3),
            GRU(200, return_sequences=False),
            Dropout(0.3),

            # Output layer
            Dense(1, activation='sigmoid')
        ])

        return model

    def create_hybrid_model(self, input_shape):
        """
        Create a hybrid model combining CNN and GRU

        Args:
            input_shape (tuple): Shape of input features
        """
        # Input layer
        inputs = Input(shape=input_shape)
        reshaped = Reshape((input_shape[0], 1))(inputs)

        # CNN branch
        cnn = Conv1D(64, 3, activation='relu', padding='same')(reshaped)
        cnn = BatchNormalization()(cnn)
        cnn = MaxPooling1D(2)(cnn)
        cnn = Conv1D(128, 3, activation='relu', padding='same')(cnn)
        cnn = BatchNormalization()(cnn)
        cnn = GlobalAveragePooling1D()(cnn)

        # GRU branch
        gru = GRU(200, return_sequences=True)(reshaped)
        gru = Dropout(0.3)(gru)
        gru = GRU(200, return_sequences=False)(gru)
        gru = Dropout(0.3)(gru)

        # Combine branches
        combined = keras.layers.concatenate([cnn, gru])

        # Final layers
        dense = Dense(128, activation='relu')(combined)
        dense = Dropout(0.5)(dense)
        dense = Dense(64, activation='relu')(dense)
        dense = Dropout(0.3)(dense)

        # Output
        outputs = Dense(1, activation='sigmoid')(dense)

        model = Model(inputs=inputs, outputs=outputs)
        return model

    def train_model(self, model, model_name, epochs=100):
        """
        Train the model with early stopping and model checkpointing

        Args:
            model: Keras model to train
            model_name (str): Name for saving the model
            epochs (int): Maximum number of epochs
        """
        print(f"Training {model_name} model...")

        # Compile model
        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='binary_crossentropy',
            metrics=['accuracy', 'precision', 'recall']
        )

        print(model.summary())

        # Callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True
        )

        model_checkpoint = ModelCheckpoint(
            f'/content/drive/MyDrive/{model_name}_best.h5',
            monitor='val_loss',
            save_best_only=True
        )

        # Train model
        history = model.fit(
            self.X_train, self.y_train,
            epochs=epochs,
            batch_size=32,
            validation_data=(self.X_val, self.y_val),
            callbacks=[early_stopping, model_checkpoint],
            verbose=1
        )

        return model, history

    def evaluate_model(self, model, model_name):
        """
        Evaluate model performance on test set

        Args:
            model: Trained Keras model
            model_name (str): Name of the model for reporting
        """
        print(f"\nEvaluating {model_name}...")

        # Predictions
        y_pred_proba = model.predict(self.X_test)
        y_pred = (y_pred_proba > 0.5).astype(int)

        # Metrics
        accuracy = accuracy_score(self.y_test, y_pred)
        auc_score = roc_auc_score(self.y_test, y_pred_proba)

        print(f"Accuracy: {accuracy:.4f}")
        print(f"AUC Score: {auc_score:.4f}")

        print("\nClassification Report:")
        print(classification_report(self.y_test, y_pred, target_names=['Mild', 'Severe']))

        print("\nConfusion Matrix:")
        cm = confusion_matrix(self.y_test, y_pred)
        print(cm)

        return {
            'accuracy': accuracy,
            'auc': auc_score,
            'predictions': y_pred,
            'probabilities': y_pred_proba,
            'confusion_matrix': cm
        }

    def plot_training_history(self, history, model_name):
        """
        Plot training history

        Args:
            history: Keras training history
            model_name (str): Name of the model
        """
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))

        # Loss
        axes[0, 0].plot(history.history['loss'], label='Training Loss')
        axes[0, 0].plot(history.history['val_loss'], label='Validation Loss')
        axes[0, 0].set_title(f'{model_name} - Loss')
        axes[0, 0].set_xlabel('Epoch')
        axes[0, 0].set_ylabel('Loss')
        axes[0, 0].legend()

        # Accuracy
        axes[0, 1].plot(history.history['accuracy'], label='Training Accuracy')
        axes[0, 1].plot(history.history['val_accuracy'], label='Validation Accuracy')
        axes[0, 1].set_title(f'{model_name} - Accuracy')
        axes[0, 1].set_xlabel('Epoch')
        axes[0, 1].set_ylabel('Accuracy')
        axes[0, 1].legend()

        # Precision
        axes[1, 0].plot(history.history['precision'], label='Training Precision')
        axes[1, 0].plot(history.history['val_precision'], label='Validation Precision')
        axes[1, 0].set_title(f'{model_name} - Precision')
        axes[1, 0].set_xlabel('Epoch')
        axes[1, 0].set_ylabel('Precision')
        axes[1, 0].legend()

        # Recall
        axes[1, 1].plot(history.history['recall'], label='Training Recall')
        axes[1, 1].plot(history.history['val_recall'], label='Validation Recall')
        axes[1, 1].set_title(f'{model_name} - Recall')
        axes[1, 1].set_xlabel('Epoch')
        axes[1, 1].set_ylabel('Recall')
        axes[1, 1].legend()

        plt.tight_layout()
        plt.savefig(f'/content/drive/MyDrive/{model_name}_training_history.png', dpi=300, bbox_inches='tight')
        plt.show()

    def compare_models(self, results_dict):
        """
        Compare performance of different models

        Args:
            results_dict (dict): Dictionary containing results from different models
        """
        print("\n" + "="*60)
        print("MODEL COMPARISON")
        print("="*60)

        comparison_df = pd.DataFrame({
            model_name: {
                'Accuracy': results['accuracy'],
                'AUC Score': results['auc']
            }
            for model_name, results in results_dict.items()
        }).T

        print(comparison_df)

        # Plot comparison
        fig, ax = plt.subplots(1, 2, figsize=(12, 5))

        models = list(results_dict.keys())
        accuracies = [results_dict[model]['accuracy'] for model in models]
        aucs = [results_dict[model]['auc'] for model in models]

        # Accuracy comparison
        ax[0].bar(models, accuracies, color=['skyblue', 'lightcoral', 'lightgreen'])
        ax[0].set_title('Model Accuracy Comparison')
        ax[0].set_ylabel('Accuracy')
        ax[0].set_ylim(0, 1)

        # AUC comparison
        ax[1].bar(models, aucs, color=['skyblue', 'lightcoral', 'lightgreen'])
        ax[1].set_title('Model AUC Comparison')
        ax[1].set_ylabel('AUC Score')
        ax[1].set_ylim(0, 1)

        plt.tight_layout()
        plt.savefig('/content/drive/MyDrive/model_comparison.png', dpi=300, bbox_inches='tight')
        plt.show()

        return comparison_df

# Usage Example and Main Execution
def main():
    """
    Main function to run the complete aphasia detection pipeline
    """
    # Initialize the system
    data_path = "/content/drive/MyDrive/Aphasia Dataset/English/Protocol/ACWT PWA/ACWT Features"
    aphasia_system = AphasiaDetectionSystem(data_path)

    # Load and preprocess data
    aphasia_system.load_data()
    aphasia_system.preprocess_data()
    aphasia_system.split_data()

    # Get input shape for models
    input_shape = (aphasia_system.X_train.shape[1],)
    print(f"Input shape for models: {input_shape}")

    # Dictionary to store results
    results = {}

    # Train and evaluate CNN model
    print("\n" + "="*60)
    print("TRAINING CNN MODEL")
    print("="*60)

    cnn_model = aphasia_system.create_cnn_model(input_shape)
    cnn_model, cnn_history = aphasia_system.train_model(cnn_model, "CNN_Aphasia")
    aphasia_system.plot_training_history(cnn_history, "CNN")
    results['CNN'] = aphasia_system.evaluate_model(cnn_model, "CNN")

    # Train and evaluate GRU model
    print("\n" + "="*60)
    print("TRAINING GRU MODEL")
    print("="*60)

    gru_model = aphasia_system.create_gru_model(input_shape)
    gru_model, gru_history = aphasia_system.train_model(gru_model, "GRU_Aphasia")
    aphasia_system.plot_training_history(gru_history, "GRU")
    results['GRU'] = aphasia_system.evaluate_model(gru_model, "GRU")

    # Train and evaluate Hybrid model
    print("\n" + "="*60)
    print("TRAINING HYBRID MODEL")
    print("="*60)

    hybrid_model = aphasia_system.create_hybrid_model(input_shape)
    hybrid_model, hybrid_history = aphasia_system.train_model(hybrid_model, "Hybrid_Aphasia")
    aphasia_system.plot_training_history(hybrid_history, "Hybrid")
    results['Hybrid'] = aphasia_system.evaluate_model(hybrid_model, "Hybrid")

    # Compare all models
    comparison_df = aphasia_system.compare_models(results)

    # Save results
    comparison_df.to_csv('/content/drive/MyDrive/model_comparison_results.csv')

    print("\n" + "="*60)
    print("PIPELINE COMPLETED SUCCESSFULLY!")
    print("="*60)
    print("All models have been trained and evaluated.")
    print("Results and plots have been saved to your Google Drive.")

    return aphasia_system, results

# Run the main function
if __name__ == "__main__":
    system, model_results = main()
```
